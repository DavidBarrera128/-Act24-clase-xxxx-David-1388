```python
class Mascota:
    def __init__(self, nombre, especie, edad, propietario):
        self.nombre = nombre
        self.especie = especie
        self.edad = edad
        self.propietario = propietario

    # --- Funciones (Acciones) ---
    def alimentar(self):
        return f"🥣 {self.nombre} ha sido alimentado/a. ¡Ya no tiene hambre!"

    def pasear(self):
        return f"🦮 {self.nombre} está dando un paseo largo con {self.propietario}."

    def vacunar(self):
        return f"💉 {self.nombre} ha sido vacunado/a. ¡Salud protegida para este {self.especie}!"

# --- Inicio del Programa Interactivo ---

print("--- REGISTRO DE MASCOTA ---")
# El usuario ingresa la información
nom = input("Nombre de la mascota: ")
esp = input("Especie (perro, gato, etc.): ")
eda = int(input("Edad (años): "))  # Convertimos a entero
pro = input("Nombre del propietario: ")

# Creamos el objeto único con esos datos
mi_mascota = Mascota(nom, esp, eda, pro)

print("\n" + "="*40)
print(f" FICHA DE IDENTIFICACIÓN: {mi_mascota.nombre.upper()} ")
print("="*40)
print(f"• Especie: {mi_mascota.especie}")
print(f"• Edad: {mi_mascota.edad} años")
print(f"• Dueño: {mi_mascota.propietario}")

# Ejecución automática de las 3 funciones solicitadas
print("\n--- INICIANDO RUTINA DIARIA ---")

mensaje_comida = mi_mascota.alimentar()
print(mensaje_comida)

mensaje_paseo = mi_mascota.pasear()
print(mensaje_paseo)

mensaje_salud = mi_mascota.vacunar()
print(mensaje_salud)

print("\n--- Tareas completadas con éxito ---")
print("programa creado por David Barrera 1388")
```
